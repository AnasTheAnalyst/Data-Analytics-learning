### Replace missing values with the average of tha coloumn

```python
# finding an average mean

mean= ks1['Age'].mean()
mean
```

```python
# replacing nan with mean

ks1['Age'] = ks1['Age'].replace(np.nan , mean)
```

## Normalization of data
- Normalization changes the value from 0 to 1(now both variable has similar influence on model)

## Method of Normalization

-- 1 Simple Feature Scaling
- x(new) =    x(old)/x(max)

- 2     Min and Maz method
- 3     Z Score (Stantard Score) -3 to +3
- 4     Log Transformation

```python
ks1[["age in days","Fare"]]
```

```python
# Simple feature Scaling

ks1["Fare"] = ks1["Fare"]/ks1["Fare"].max()
ks1.head()
```

```python
# Min - Max Method
ks1["Fare"] = (ks1["Fare"] - ks1["Fare"].min())/(ks1["Fare"].max()-ks1["Fare"].min() )
ks1.head()
```

```python
ks1[["age in days","Fare"]]
```

```python
# Z score method

ks1["Fare"] = (ks1["Fare"] - ks1["Fare"].mean())/(ks1["Fare"].std() )
```

```python
ks1[["age in days","Fare"]]
```

```python
# log transformation

ks["Fare"] = np.log (ks["Fare"])
ks.head()
```
