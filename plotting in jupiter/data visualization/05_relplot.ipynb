```python
import seaborn as sns

df =sns.load_dataset("titanic")

df.head()
```

```python
sns.set_style('white')

sns.relplot(x='age' ,y='fare',data=df ,hue='pclass'        # use in scatter plot
            ,style='pclass' ,kind='line')
```

# col_Wrap

```python
sns.relplot(x='age' ,y='fare',data=df,
           col='deck')
```

```python
sns.relplot(x='age' ,y='fare',data=df,
           col='deck',
           col_wrap=3)
```

# styling

```python
sns.relplot(x='age' ,y='fare',data=df,
           col='pclass',
           hue="survived",
           s=150,
           palette=["red","black"],
           marker='*',    style="survived")
```

# marker and sizes

```python
sns.relplot(x='age' ,y='fare',data=df,
           col='pclass',
           hue="survived",
           s=150,
           palette=["red","black"],    style="survived",     markers=['X','o'])
```

# LINE PLOT

```python
sns.relplot(x='deck' ,y='survived',         data=df[df.age<70],
            kind="line",
           hue="sex")
```

```python
g =sns.relplot(x='age' ,y='survived',         data=df[df.age<70],
            kind="line",
           hue="sex",
           ci=None)

g.set_xlabels("number of people survived")
g.set_ylabels("people survived")
```
