# other method (we can also uses data on x axis and y axis)

```python
g =sns.FacetGrid(penguin , col="island")
g.map_dataframe(sns.histplot ,y="flipper_length_mm")
```

```python
g =sns.FacetGrid(penguin , col="island")
g.map_dataframe(sns.scatterplot , x='bill_depth_mm', y='bill_length_mm')
g.set_axis_labels('depth','length')
```

```python
# STYLING THE TITLE NAME and row

g =sns.FacetGrid(penguin , col="island", row="species")
g.map_dataframe(sns.scatterplot , x='bill_depth_mm' ,y='bill_length_mm')

g.set_axis_labels('depth','length')
g.set_titles(col_template='{col_name} island')
```

## share a coloumn on y axis (sharey=False)

```python
g =sns.FacetGrid(penguin , col="island",sharey=False,  ylim=(20,60))
g.map_dataframe(sns.scatterplot , x='bill_depth_mm' ,y='bill_length_mm')
g.set_axis_labels('depth','length')
g.set_titles(col_template='{col_name} island')
```

```python
# Adding hue and sns.add_legend

g =sns.FacetGrid(penguin , col="island", sharey=False, ylim=(20,60),   hue="species")

g.map_dataframe(sns.scatterplot , x='bill_depth_mm' ,y='bill_length_mm' )
g.set_axis_labels('depth','length')
g.set_titles(col_template='{col_name} island')
g.add_legend()
```

```python
# Adding Palette

g =sns.FacetGrid(penguin , col="island", sharey=False, ylim=(20,60),   hue="species", palette="prism")

g.map_dataframe(sns.scatterplot , x='bill_depth_mm' ,y='bill_length_mm' )
g.set_axis_labels('depth','length')
g.set_titles(col_template='{col_name} island')
g.add_legend()
```

# Custom funtion

```python
import matplotlib.pyplot as plt
import numpy as np

def add_mean_line(data,var=None ,**kws):
    
    # if no variable provided skip adding mean line
    
    if not var:return
    
    #calculate mean for each group
    
    m=np.mean(data[var])
    
    # get current axis
    ax = plt.gca()
    
    # add group line at group mean
    ax.axvline(m,  color='maroon' ,  lw =3, ls ='--')
    
    # annovate group mean
    x_pos=0.65
    if m >5000:x_pos=0.2 
    ax.text(x_pos,0.7, 'mean=(m=.0)',
            transform=ax.transAxes ,         #tranform position of range to(0,0) to(1,1)
            color='maroon' ,fontweight='bold', fontsize=12)
    
```

```python
# Custom funtion

quantity='bill_length_mm'

g =sns.FacetGrid(penguin ,row="sex" ,col="island" )
g.map_dataframe(sns.kdeplot , x=quantity , lw=3 , shade=True )

g.map_dataframe(add_mean_line,  var=quantity) # add a "bill_length_mm " here

g.set_titles(row_template='{row_name}',col_template='{col_name} penguins')



```
